```python
# 疑似データの生成

import numpy as np

n, p = 50, 200
X = np.random.randn(n, p)
y = 3*X[:,0] - 2*X[:,1] + X[:,2] + np.random.randn(n)
```

```python
# Lasso + クロスバリデーションによる回帰係数の推定
# 参考: https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LassoCV.html

from sklearn.linear_model import LassoCV

reg = LassoCV(cv=5).fit(X, y)
print(reg.coef_)
```
